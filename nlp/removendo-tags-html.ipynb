```python
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def strip_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    [s.extract() for s in soup(['iframe', 'script'])]
    stripped_text = soup.get_text()
    stripped_text = re.sub(r'[\r|\n|\n\r]+', '\n', stripped_text)
    return stripped_text
```

```python
df = pd.read_excel('voto-processo.xlsx')
df
```

```python
# df.iloc[<linhas>, <colunas>]
df2 = df.iloc[0:1,2:3]
```

```python
texto = df2['conteudo'].values[0]
texto
```

```python
clean_content = strip_html_tags(texto)
clean_content
```
